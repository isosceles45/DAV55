```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>D3.js Event Handling</title>
  <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<h2>Event Handling in D3</h2>
<div id="chart-container"></div>

<script>
const data = [
  { x: 10, y: 20 },
  { x: 50, y: 70 },
  { x: 90, y: 50 },
];

const svgWidth = 200;
const svgHeight = 300;
const margin = { top: 20, right: 30, bottom: 30, left: 40 };
const width = svgWidth - margin.left - margin.right;
const height = svgHeight - margin.top - margin.bottom;

const svg = d3.select("#chart-container")
  .append("svg")
  .attr("width", svgWidth)
  .attr("height", svgHeight)
  .append("g")
  .attr("transform", `translate(${margin.left},${margin.top})`);

const xScale = d3.scaleLinear()
  .domain([0, d3.max(data, d => d.x)])
  .range([0, width]);

const yScale = d3.scaleLinear()
  .domain([0, d3.max(data, d => d.y)])
  .range([height, 0]);

svg.selectAll(".circle")
  .data(data)
  .enter().append("circle")
  .attr("class", "circle")
  .attr("cx", d => xScale(d.x))
  .attr("cy", d => yScale(d.y))
  .attr("r", 5)
  .attr("fill", "steelblue")
  .on("mouseover", handleMouseOver)
  .on("mouseout", handleMouseOut);

function handleMouseOver(d, i) {
  d3.select(this)
    .transition()
    .duration(100)
    .attr("r", 8)
    .attr("fill", "orange");
}

function handleMouseOut(d, i) {
  d3.select(this)
    .transition()
    .duration(100)
    .attr("r", 5)
    .attr("fill", "steelblue");
}

</script>

</body>
</html>
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Interactive Drawing Tool with D3.js</title>
    <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>

    <h2>Interactive Drawing Tool with D3.js</h2>

    <svg id="drawingCanvas" width="800px" height="600px">
        <rect width="100%" height="100%" fill="white"></rect>
    </svg>

    <br />
    <button id="clearButton">Clear Canvas</button>

    <script>

        const svg = d3.select("#drawingCanvas");
        let isDrawing = false;
        let pathData = "";

        svg.on("mousedown", () => {
            isDrawing = true;
            pathData = "";
            const pos = d3.mouse(svg.node());
            pathData += `M ${pos[0]},${pos[1]}`;
        });

        svg.on("mousemove", () => {
            if (isDrawing) {
                const pos = d3.mouse(svg.node());
                pathData += ` L ${pos[0]},${pos[1]}`;
                svg.selectAll("path").remove();
                svg.append("path")
                    .attr("d", pathData)
                    .attr("stroke", "black")
                    .attr("stroke-width", 2)
                    .attr("fill", "none");
            }
        });

        svg.on("mouseup", () => {
            isDrawing = false;
        });

        d3.select("#clearButton").on("click", () => {
            svg.selectAll("path").remove();
        });

    </script>

</body>
</html>
```
