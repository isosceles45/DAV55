```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Data-Based D3.js Example</title>
    <script src="https://d3js.org/d3.v7.min.js"></script> <!-- Updated to D3 version 7 -->
</head>
<body>

    <h2>Modifying Elements</h2>
    <button id="random-button">Add Random Colorful Text</button>
    <div id="chart-container"></div>

    <script>
      const chartContainer = d3.select('#chart-container');
      let count = 0;

      d3.select("#random-button").on("click", function() {
          chartContainer
            .append('p')
            .text(`Text number ${count++}`)
            .style('color', randomColors());
      });

      function randomColors() {
        const letters = "0123456789ABCDEF";
        let color = "#";
        for (let i = 0; i < 6; i++) {
          color += letters[Math.floor(Math.random() * 16)];
        }
        return color;
      }
    </script>

</body>
</html>
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>D3.js Line Chart</title>
  <script src="https://d3js.org/d3.v5.min.js"></script>
</head>
<body>
<h2>Data Loading in D3</h2>
<div id="chart-container"></div>

<script>
const data = [
  { category: "A", value: 100 },
  { category: "B", value: 200 },
  { category: "C", value: 150 },
];

const svgWidth = 400;
const svgHeight = 300;
const margin = { top: 20, right: 30, bottom: 30, left: 40 };
const width = svgWidth - margin.left - margin.right;
const height = svgHeight - margin.top - margin.bottom;

const svg = d3.select("#chart-container")
  .append("svg")
  .attr("width", svgWidth)
  .attr("height", svgHeight)
  .append("g")
  .attr("transform", `translate(${margin.left},${margin.top})`);

const x = d3.scaleBand()
  .range([0, width])
  .domain(data.map(d => d.category))
  .padding(0.1);

const y = d3.scaleLinear()
  .domain([0, d3.max(data, d => d.value)])
  .range([height, 0]);

svg.append("g")
  .attr("transform", `translate(0,${height})`)
  .call(d3.axisBottom(x));

svg.append("g")
  .call(d3.axisLeft(y));

svg.selectAll(".line")
  .data([data])
  .enter().append("path")
  .attr("class", "line")
  .attr("d", d3.line()
    .x(d => x(d.category) + x.bandwidth() / 2)
    .y(d => y(d.value))
  )
  .attr("fill", "none")
  .attr("stroke", "steelblue")
  .attr("stroke-width", 1.5);

svg.selectAll(".dot")
  .data(data)
  .enter().append("circle")
  .attr("class", "dot")
  .attr("cx", d => x(d.category) + x.bandwidth() / 2)
  .attr("cy", d => y(d.value))
  .attr("r", 5)
  .attr("fill", "steelblue");

</script>

</body>
</html>
```
